```python
import os
import subprocess
import sys

def run(cmd):
    print(f"\n$ {cmd}")
    subprocess.run(cmd, shell=True, check=True)

py = f'"{sys.executable}"'
os.environ["MPLCONFIGDIR"] = "/tmp/mpl"

run(f"{py} -m pip install -r requirements.txt")
run("chmod +x scripts/download_dataset.sh")
run("./scripts/download_dataset.sh")
run(f"{py} scripts/prepare_pretrain_dataset.py --data data/diabetes_prediction_dataset.csv")
run(f"{py} scripts/train_model.py --data data/processed/diabetes_train.csv --epochs 30 --batch-size 128")
run(f"{py} scripts/evaluate_model.py --data data/processed/diabetes_test.csv")

print("\nTraining complete.")
print("Model artifacts folder: models/")
print("Processed dataset folder: data/processed/")
```
